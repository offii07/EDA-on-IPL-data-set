# What is the percentage of winning the toss and then winning the match ?

```python
# Finding out how many times the team won after winning the toss.
toss_win_and_match_win=np.sum(ipl['toss_winner'] == ipl['winner'])
toss_win_and_match_win
```

```python
# Total number of matches played in the ipl.
total_matches_played=756
total_matches_played
```

```python
# Winning percentage of the match after toss won.
toss_win_and_match_win/total_matches_played*100
```
